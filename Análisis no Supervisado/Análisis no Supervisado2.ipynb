```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
file_path = r'C:\Users\trist\Downloads\comprar_alquilar.csv'
data = pd.read_csv(file_path)
```

```python
X = data.drop(columns=['comprar'])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
pca_df = pd.DataFrame(data=X_pca, columns=['Componente Principal 1', 'Componente Principal 2'])
```

```python
pca_df['comprar'] = data['comprar']
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(pca_df['Componente Principal 1'], pca_df['Componente Principal 2'], c=pca_df['comprar'], cmap='viridis')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Proyección de PCA en un plano bidimensional')
plt.colorbar(label='Etiqueta "Comprar"')
plt.show()
```
